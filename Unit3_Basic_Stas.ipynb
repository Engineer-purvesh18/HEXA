### Q1

```python
# using statistics module for mean/median/mode
from statistics import mean, median, mode, multimode

data = [12, 15, 12, 14, 19, 16, 18, 12, 16, 15]
print("Mean:", mean(data))
print("Median:", median(data))
print("Mode:", multimode(data))
```

### Q2

```python
# sample variance/std since this is test scores from a class (a sample, not full population)
import numpy as np

scores = [45, 55, 65, 75, 85, 95, 100, 85, 90, 75]
scores = np.array(scores)

variance = np.var(scores, ddof=1)   # sample variance
std_dev = np.std(scores, ddof=1)    # sample std dev

print("Variance (sample):", variance)
print("Standard Deviation (sample):", std_dev)
```

### Q3

```python
# weekly sales numbers
import numpy as np

sales = [50, 60, 55, 65, 60, 70]
sales = np.array(sales)

data_range = sales.max() - sales.min()
q1 = np.percentile(sales, 25)
q3 = np.percentile(sales, 75)
iqr = q3 - q1

print("Range:", data_range)
print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)
```

### Q5

```python
# using z-score + normal cdf from scipy
from scipy import stats

mean_height = 170
std_height = 10

z = (180 - mean_height) / std_height
prob = 1 - stats.norm.cdf(z)

print("Z-score:", z)
print("P(height > 180):", prob)
```
