## Case 1: Teaching Methods

### Q1

```python
# group-wise mean/median/std
import numpy as np

group1 = [65, 70, 75, 80, 85, 78, 82, 79, 74, 68, 72, 77, 71, 84, 73]
group2 = [58, 62, 67, 65, 69, 70, 64, 60, 66, 68, 71, 59, 63, 67, 61]
group3 = [75, 78, 80, 85, 87, 82, 79, 88, 83, 81, 85, 84, 89, 86, 80]

for name, g in [("Group 1 (Traditional)", group1), ("Group 2 (Online)", group2), ("Group 3 (Blended)", group3)]:
    g = np.array(g)
    print(f"{name}: mean={g.mean():.2f}, median={np.median(g):.2f}, std={g.std(ddof=1):.2f}")
```

### Q2

```python
# one-way ANOVA using scipy
from scipy import stats

group1 = [65, 70, 75, 80, 85, 78, 82, 79, 74, 68, 72, 77, 71, 84, 73]
group2 = [58, 62, 67, 65, 69, 70, 64, 60, 66, 68, 71, 59, 63, 67, 61]
group3 = [75, 78, 80, 85, 87, 82, 79, 88, 83, 81, 85, 84, 89, 86, 80]

f_stat, p_value = stats.f_oneway(group1, group2, group3)
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Result: statistically significant difference between the groups (reject H0)")
else:
    print("Result: no statistically significant difference (fail to reject H0)")
```

### Q3

```python
# tukey hsd for pairwise comparisons since anova came back significant
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd

group1 = [65, 70, 75, 80, 85, 78, 82, 79, 74, 68, 72, 77, 71, 84, 73]
group2 = [58, 62, 67, 65, 69, 70, 64, 60, 66, 68, 71, 59, 63, 67, 61]
group3 = [75, 78, 80, 85, 87, 82, 79, 88, 83, 81, 85, 84, 89, 86, 80]

scores = group1 + group2 + group3
labels = ["Traditional"]*len(group1) + ["Online"]*len(group2) + ["Blended"]*len(group3)

tukey = pairwise_tukeyhsd(endog=scores, groups=labels, alpha=0.05)
print(tukey)
```

### Q4

```python
# normality + equal variance checks
from scipy import stats

group1 = [65, 70, 75, 80, 85, 78, 82, 79, 74, 68, 72, 77, 71, 84, 73]
group2 = [58, 62, 67, 65, 69, 70, 64, 60, 66, 68, 71, 59, 63, 67, 61]
group3 = [75, 78, 80, 85, 87, 82, 79, 88, 83, 81, 85, 84, 89, 86, 80]

for name, g in [("Group 1", group1), ("Group 2", group2), ("Group 3", group3)]:
    stat, p = stats.shapiro(g)
    print(f"Shapiro-Wilk {name}: stat={stat:.4f}, p={p:.4f}")

levene_stat, levene_p = stats.levene(group1, group2, group3)
print(f"\nLevene's test: stat={levene_stat:.4f}, p={levene_p:.4f}")
```

**Q5.** Blended learning comes out on top - highest mean (82.80) and it's significantly different from both other groups in the Tukey test. Traditional is second (75.53), and online learning is lowest (64.67). Normality and equal-variance assumptions both hold (all Shapiro p-values > 0.05, Levene's p = 0.19), so the ANOVA result can be trusted. if I had to recommend one method, it'd be blended learning.

## Case 2: Marketing Campaigns

### Q1

```python
# region-wise mean/median/std
import numpy as np

north = [102, 110, 108, 104, 112, 109, 115, 114, 120, 118, 107, 113, 117, 121, 116, 111, 119, 122, 125, 126]
south = [95, 100, 98, 102, 97, 101, 104, 105, 108, 106, 99, 103, 107, 109, 110, 98, 102, 108, 105, 107]
west  = [110, 115, 120, 125, 123, 119, 121, 124, 127, 130, 118, 126, 129, 128, 131, 117, 122, 128, 132, 134]

for name, g in [("North", north), ("South", south), ("West", west)]:
    g = np.array(g)
    print(f"{name}: mean={g.mean():.2f}, median={np.median(g):.2f}, std={g.std(ddof=1):.2f}")
```

### Q2

```python
# one-way ANOVA across regions
from scipy import stats

north = [102, 110, 108, 104, 112, 109, 115, 114, 120, 118, 107, 113, 117, 121, 116, 111, 119, 122, 125, 126]
south = [95, 100, 98, 102, 97, 101, 104, 105, 108, 106, 99, 103, 107, 109, 110, 98, 102, 108, 105, 107]
west  = [110, 115, 120, 125, 123, 119, 121, 124, 127, 130, 118, 126, 129, 128, 131, 117, 122, 128, 132, 134]

f_stat, p_value = stats.f_oneway(north, south, west)
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Result: significant difference in sales between regions (reject H0)")
else:
    print("Result: no significant difference (fail to reject H0)")
```

### Q3

```python
# tukey hsd - anova was significant so checking which regions differ
from statsmodels.stats.multicomp import pairwise_tukeyhsd

north = [102, 110, 108, 104, 112, 109, 115, 114, 120, 118, 107, 113, 117, 121, 116, 111, 119, 122, 125, 126]
south = [95, 100, 98, 102, 97, 101, 104, 105, 108, 106, 99, 103, 107, 109, 110, 98, 102, 108, 105, 107]
west  = [110, 115, 120, 125, 123, 119, 121, 124, 127, 130, 118, 126, 129, 128, 131, 117, 122, 128, 132, 134]

sales = north + south + west
labels = ["North"]*len(north) + ["South"]*len(south) + ["West"]*len(west)

tukey = pairwise_tukeyhsd(endog=sales, groups=labels, alpha=0.05)
print(tukey)
```

### Q4

```python
# normality + equal variance checks for the regions
from scipy import stats

north = [102, 110, 108, 104, 112, 109, 115, 114, 120, 118, 107, 113, 117, 121, 116, 111, 119, 122, 125, 126]
south = [95, 100, 98, 102, 97, 101, 104, 105, 108, 106, 99, 103, 107, 109, 110, 98, 102, 108, 105, 107]
west  = [110, 115, 120, 125, 123, 119, 121, 124, 127, 130, 118, 126, 129, 128, 131, 117, 122, 128, 132, 134]

for name, g in [("North", north), ("South", south), ("West", west)]:
    stat, p = stats.shapiro(g)
    print(f"Shapiro-Wilk {name}: stat={stat:.4f}, p={p:.4f}")

levene_stat, levene_p = stats.levene(north, south, west)
print(f"\nLevene's test: stat={levene_stat:.4f}, p={levene_p:.4f}")
```

### Q5

```python
# one-sample t-test against the $110k target for each region
from scipy import stats

north = [102, 110, 108, 104, 112, 109, 115, 114, 120, 118, 107, 113, 117, 121, 116, 111, 119, 122, 125, 126]
south = [95, 100, 98, 102, 97, 101, 104, 105, 108, 106, 99, 103, 107, 109, 110, 98, 102, 108, 105, 107]
west  = [110, 115, 120, 125, 123, 119, 121, 124, 127, 130, 118, 126, 129, 128, 131, 117, 122, 128, 132, 134]

for name, g in [("North", north), ("South", south), ("West", west)]:
    t_stat, p_value = stats.ttest_1samp(g, 110)
    print(f"{name}: mean={sum(g)/len(g):.2f}, t={t_stat:.4f}, p={p_value:.4f}")
```

**Q6.** West region has the highest average sales (123.95k) and it's significantly higher than both North (114.45k) and South (103.20k) per Tukey. North does clear the $110k bar (t=2.98, p=0.008), West obviously does too (t=10.02, p<0.001) - but South falls significantly short of $110k (t=-6.94, p<0.001), so that campaign isn't meeting expectations. West's campaign looks like the most effective one, and South's needs a rethink since it's not just missing the target, it's missing it by a clear margin.
