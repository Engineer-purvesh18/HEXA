### Q2

```python
# generating 10 random numbers between 1 to 100
import random
random.seed(7)
numbers = [random.randint(1, 100) for _ in range(10)]
print("Original list:", numbers)
print("Ascending:", sorted(numbers))
print("Descending:", sorted(numbers, reverse=True))
print("Minimum:", min(numbers))
print("Maximum:", max(numbers))
print("Average:", sum(numbers) / len(numbers))
```

```python
# 1D array 1 to 10
import numpy as np

arr = np.arange(1, 11)
print("Original array:", arr)

reshaped = arr.reshape(2, 5)
print("Reshaped (2x5):\n", reshaped)

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard deviation:", np.std(arr))

print("Elements at even indices:", arr[::2])
```

### Q7

```python
# creating a dummy csv first since none was given
import pandas as pd

# create a sample csv to work with (in practice this would already exist)
sample = pd.DataFrame({
    "Product": ["Pen", "Notebook", "Bag", "Bottle", "Pencil", "Eraser"],
    "Price": [10, 40, 500, 150, 5, 3],
    "Quantity": [100, 50, 20, 30, 200, 150]
})
sample.to_csv("sales.csv", index=False)

# Read data from CSV into a DataFrame
df = pd.read_csv("sales.csv")

print("First five rows:\n", df.head())

# Modify a column - add 10% tax to Price
df["Price"] = df["Price"] * 1.10

# Write back to a new CSV
df.to_csv("sales_updated.csv", index=False)
print("\nUpdated data written to sales_updated.csv:\n", df)
```

### Q8

```python
# sample employee data
import pandas as pd

data = {
    "Department": ["HR", "IT", "Finance", "HR", "IT"],
    "Employee": ["John", "Anna", "Peter", "Mike", "Linda"],
    "Salary": [50000, 62000, 48000, 52000, 58000]
}
df = pd.DataFrame(data)

grouped = df.groupby("Department")["Salary"].mean()
print("Average salary by department:\n", grouped)
```

### Q9

```python
# sample dates
import pandas as pd

df = pd.DataFrame({
    "Date": ["2023-01-15", "2023-03-22", "2023-07-08", "2023-11-30"]
})
df["Date"] = pd.to_datetime(df["Date"])

df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day

print(df)

date_diff = df["Date"].iloc[3] - df["Date"].iloc[0]
print("\nDifference between first and last date:", date_diff)
```

### Q10

```python
# dataframe with some NaN values on purpose
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A": [10, 20, np.nan, 40, 50],
    "B": [5, np.nan, 15, 20, np.nan],
    "C": [1, 2, 3, 4, 5]
})
print("Original DataFrame:\n", df)

print("\nMissing values per column:\n", df.isnull().sum())

df_filled = df.fillna(df.mean(numeric_only=True))
print("\nAfter filling missing values with column mean:\n", df_filled)

df_dropped = df.dropna()
print("\nAfter dropping rows with any remaining missing values:\n", df_dropped)
```
